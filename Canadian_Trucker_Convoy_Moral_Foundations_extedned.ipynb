1. Use All Probabilities per Word and Return Sentiment Scores

This option should be used when one wants to extract the overall, holistic moral signal from a document. Note that because each word is assigned five foundation probabilities, there exist higher correlations across these foundations, making this method less suitable when one wants to

use the foundation probabilities as predictor variables in statistical models discriminate which foundations are more or less represented in a text. For these cases, options (2) and (4) below should be preferred.

```python
from emfdscore.scoring import score_docs 

num_docs = len(template_input)

DICT_TYPE = 'emfd'
PROB_MAP = 'all'
SCORE_METHOD = 'bow'
OUT_METRICS = 'sentiment'
OUT_CSV_PATH = 'all-sent.csv'

df = score_docs(template_input,DICT_TYPE,PROB_MAP,SCORE_METHOD,OUT_METRICS,num_docs)
df.to_csv(OUT_CSV_PATH, index=False)
```

```python
# Inspect output 
all_sent = pd.read_csv('all-sent.csv')
all_sent.head()
```

As the above dataframe illustrates, each document is assigned 5 foundation probabilities that denote the average probabilty of each document belonging to one of the five moral foundations and 5 sentiment scores that describe the average sentiment of detected moral words for that foundation.

In addition, eMFDscore always returns the detected moral-to-nonmoral word ratio(moral_nonmoral_ratio), along with the variance across the foundation scores (f_var) as well as the variance across the sentiment scores (sent_var).

2. Assign Single Probability per Word and Return Sentiment Scores

```python
from emfdscore.scoring import score_docs 

num_docs = len(template_input)

DICT_TYPE = 'emfd'
PROB_MAP = 'single'
SCORE_METHOD = 'bow'
OUT_METRICS = 'sentiment'
OUT_CSV_PATH = 'single-sent.csv'

df = score_docs(template_input,DICT_TYPE,PROB_MAP,SCORE_METHOD,OUT_METRICS,num_docs)
df.to_csv(OUT_CSV_PATH, index=False)
```

```python
# Inspect output 
single_sent = pd.read_csv('single-sent.csv')
single_sent.head()
```

Here, the returned output is the same as in option (1) above, although every word was assigned only one probability score when scoring the document.

# Merge with the original dataframe

```python
df_merged = pd.read_csv('/content/drive/MyDrive/Social Network Analysis_Convoy_Canada/word2vec_completed_to_MFA.csv')
```

```python
df_merged_all_vv = all_vv.merge(df_merged, left_index=True, right_index=True)
```

```python
df_merged_all_vv
```

```python
emfd_virtue = df_merged_all_vv.groupby(df_merged_all_vv.labels).mean()
emfd_virtue['labels'] = emfd_virtue.index
emfd_virtue = emfd_virtue[['labels', 'care.virtue', 'fairness.virtue', 'loyalty.virtue', 'authority.virtue', 'sanctity.virtue']]
emfd_virtue
```

```python
f, ax = plt.subplots(figsize=(7, 5))
pd.plotting.parallel_coordinates(emfd_virtue, 'labels', 
                                 cols=['care.virtue', 'fairness.virtue', 'loyalty.virtue', 'authority.virtue', 'sanctity.virtue'])
ax.legend(loc="lower left");
```
